```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm','Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})

df
```

## Question 1

```python
df1 = df
for i in range(len(df1)):
    if str(df1['FlightNumber'][i]) == str('nan'):
        df1['FlightNumber'][i] = df1['FlightNumber'][i-1]+10
df1['FlightNumber'] = df1['FlightNumber'].astype(int)
print(df1.dtypes)
df1
```

## Question 2

```python
df2 = pd.DataFrame(columns=['From','To'])
df2['From'] = df1['From_To'].apply(lambda x: x.split('_')[0])
df2['To'] = df1['From_To'].apply(lambda x: x.split('_')[1])
df2
```

## Question 3

```python
df2['From'] = df2['From'].apply(lambda x: x.capitalize())
df2['To'] = df2['To'].apply(lambda x: x.capitalize())
df2
```

## Question 4

```python
df1.drop(columns=['From_To'],inplace=True)
df3 = pd.concat([df1,df2],axis=1)
df3
```

## Question 5

```python
recentDelays = df3[['RecentDelays']]
recentDelays
```

```python
lst = []
for i in range(len(recentDelays['RecentDelays'])):
    lst.append(recentDelays['RecentDelays'].iloc[i])
lst
```

```python
delaysdf = pd.DataFrame(lst, columns=[f'delay_{i}' for i in range(3)])
delaysdf
```

```python
df_comb = pd.concat([df3,delaysdf], axis=1)
df_comb
```

```python
final_df = df_comb.drop(columns='RecentDelays')
final_df
```
