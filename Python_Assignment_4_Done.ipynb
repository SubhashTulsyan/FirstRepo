## Question 1.1

```python
class Triangle:
    def __init__(self, a, b, c):
        self.a = a
        self.b = b
        self.c = c

class Area(Triangle):
    def __init__(self, *args):
        super(Area, self).__init__(*args)
    
    def getArea(self):
        s = (self.a+self.b+self.c)/2
        area = (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5
        return area
        
area = Area(10,20,12)
print(area.getArea())
```

## Question 1.2

```python
def filter_long_words(lst: list, n: int):
    res = []
    for i in lst:
        if len(i)>n:
            res.append(i)
    return res

res = filter_long_words(['fnbsdkjsdnjkngklns','ee','dksgbsdknb'],9)
res
```

## Question 2.1

```python
def lenOfWords(lst: list):
    return list(map(lambda x: len(x), lst))
lenOfWords(['ab','cde','erty','ksbgnsjkgnjgnsk'])
```

## Question 2.2

```python
def isVowel(a: str):
    if len(a)==1 and a in ['a','e','i','o','u']:
        return True
    return False
print(isVowel('p'))
print(isVowel('o'))
```
