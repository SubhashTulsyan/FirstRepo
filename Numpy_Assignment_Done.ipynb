## Question 1

```python
import numpy as np

def asd(a: list, k=4):
    lst = []
    lst1 = []
    for i in a:
        for j in range(1,k+1):
            lst1.append(i**(j-1))
        lst.append(lst1)
        lst1 = []
    return np.array(lst)
            
asd([1,2,3,4,5],k=7)
```

## Question 2

```python
def mov_ave(arr,k):
    l1 = []
    n = len(arr)
    for i in range(n-k+1):
        l1.append(sum(arr[i:i+k])/k)
    return l1


arr2 = np.array([3, 5, 7, 2, 8, 10, 11, 65, 72, 81, 99, 100, 150])
mov_ave(arr2,k=3)
```
