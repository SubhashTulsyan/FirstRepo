## Question 1.1

```python
def myReduce(fn,iterobj):
    s = iterobj[0]
    for i in range(len(iterobj)-1):
        s = fn(s,iterobj[i+1])
    return s

fn = lambda x,y: x+y
l2 = [1,2,3,4,5,6,7,8,11]

myReduce(fn,l2)
```

## Question 1.2

```python
def myFilter(fn,iterobj):
    res = []
    for i in iterobj:
        if fn(i):
            res.append(i)
    return res

def fn(x):
    if x%2==0:
        return True
    return False

l2 = [1,2,3,4,5,6,7,8,11]
myFilter(fn,l2)
```
